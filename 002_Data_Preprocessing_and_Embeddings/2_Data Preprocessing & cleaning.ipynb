###Chat Conversion Handle

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible'
}


{
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
def remove_stopwords(text):
  new_text = []

  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

### Remove Emoji Handle

```python
import re
```

```python
def remove_emoji(text):
  emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', text)
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import spacy
nlp = spacy.load("en_core_web_sm")


nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
# 2.1 Data Cleanup


# Handling Missing Values
print("Handling Missing Values")
data = {'Name': ['Alice', 'Bob', np.nan, 'David'], 'Age': [25, np.nan, 30, 40]}
df = pd.DataFrame(data)
print(df)
df['Age'].fillna(df['Age'].mean(), inplace=True)  # Imputation
print(df)
df.dropna(subset=['Name'], inplace=True)  # Removing missing values
print(df)
print('-'*200)


# Noise Removal
print("Noise Removal")
def correct_spelling(text):
    corrections = {'hte': 'the', 'recieve': 'receive'}
    return ' '.join([corrections.get(word, word) for word in text.split()])

print("hte cat sat on hte mat")
print(correct_spelling("hte cat sat on hte mat"))  # Example
print('-'*200)


# Duplicate Removal
print("Duplicate Removal")
data = {'Product': ['Apple', 'Banana', 'Apple'], 'Price': [100, 50, 100]}
df = pd.DataFrame(data)
print(df)
df.drop_duplicates(inplace=True)
print(df)
print('-'*200)



# 2.2 Basic Preprocessing


# Tokenization
print("Tokenization")
sentence = "The cat sat on the mat. The dog barked."
print(sentence)
print(sent_tokenize(sentence))
print(word_tokenize(sentence))
print('-'*200)


# Text Normalization
print("Text Normalization")
def normalize_text(text):
    return text.encode('utf-8', 'ignore').decode('utf-8')

text = "Hello, ä¸–ç•Œ"
print(text)
normalize_text(text)
print('-'*200)





# 2.3 Optional Preprocessing


# Stop Word Removal
print("Stop Word Removal")
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = word_tokenize(text)
    return ' '.join([word for word in words if word.lower() not in stop_words])

print("The cat sat on a mat")
print(remove_stopwords("The cat sat on a mat"))
print('-'*200)


# Stemming
print("Stemming")
stemmer = PorterStemmer()
def stem_text(text):
    words = word_tokenize(text)
    return ' '.join([stemmer.stem(word) for word in words])

print("running runs ran")
print(stem_text("running runs ran"))
print('-'*200)


# Lemmatization
print("Lemmatization")
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    words = word_tokenize(text)
    return ' '.join([lemmatizer.lemmatize(word) for word in words])
print("running better")
print(lemmatize_text("running better"))
print('-'*200)


# Punctuation Removal
print("Punctuation Removal")
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

print("Hello, world!")
print(remove_punctuation("Hello, world!"))
print('-'*200)


# Lowercasing
print("Lowercasing")
def to_lowercase(text):
    return text.lower()

print("The Cat Sat")
print(to_lowercase("The Cat Sat"))
print('-'*200)


# Language Detection
print("Language Detection")
import langdetect
from langdetect import detect

def detect_language(text):
    return detect(text)

print("Bonjour le monde")
print(detect_language("Bonjour le monde"))
print('-'*200)


# 2.4 Advanced Preprocessing
print("Advanced Preprocessing")

print('-'*200)


# Part-of-Speech Tagging
print("Part-of-Speech Tagging")
def pos_tagging(text):
    doc = nlp(text)
    return [(token.text, token.pos_) for token in doc]

print("The cat sat on the mat")
print(pos_tagging("The cat sat on the mat"))
print('-'*200)


# Parsing
print("Parsing")
def parse_sentence(text):
    doc = nlp(text)
    return [(token.text, token.dep_) for token in doc]
print("The cat sat on the mat")
print(parse_sentence("The cat sat on the mat"))
print('-'*200)


# Coreference Resolution (Simplified)
print("Coreference Resolution (Simplified)")
def coreference_resolution(text):
    doc = nlp(text)
    return [(token.text, token.head.text) for token in doc if token.dep_ == "nsubj"]

print("John went to the store. He bought a book.")
print(coreference_resolution("John went to the store. He bought a book."))
print('-'*200)
```
