### Punctuation handling

```python
import string, time
```

```python
string.punctuation
```

```python
punc = string.punctuation
punc
```

```python
def remove_punc(text):
  for char in punc:
    text = text.replace(char,'')
  return text
```

```python
text = "Hello, World!"
```

```python
start = time.time()
print(remove_punc(text))
time1 = time.time() - start
print(time1*50000)
```

```python
def remove_punc1(text):
  return text.translate(str.maketrans('','',punc))
```

```python
start = time.time()
print(remove_punc1(text))
time2 = time.time() - start
print(time2*50000)
```

### Chat Conversion Handle

```python
chat_words = {
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion("you doing really well, OMG")
```

```python
len(stopwords.words('english'))
```

```python
def remove_stopwords(text):
  new_text = []
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x = new_text[:]
  new_text.clear()
  return " ".join(x)
```

###Remove Emoji handle

```python
import re
```

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

###Stemmer

A **stemmer** is a tool used in text processing to simplify words by removing their endings or prefixes, reducing them to their root form. For example, it turns words like *running*, *runner*, and *ran* into the base form *run*.

The main purpose of stemming is to make it easier for computers to analyze and compare words, especially in search engines or natural language processing tasks. By reducing words to their root, a stemmer helps group related words together and makes processing text faster and simpler.

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

###Lemmatization

**Lemmatization** is a process in text analysis that simplifies a word to its base form, called the **lemma**, but it does so in a way that keeps the word meaningful in its context. Unlike stemming, which just chops off word endings, lemmatization uses a dictionary and grammar rules to find the correct base form.

For example:
- The words *running*, *ran*, and *runs* are all lemmatized to *run* because they are forms of the same verb.
- The word *better* would be lemmatized to *good* (its base form) because lemmatization understands grammar and meaning.

In short, lemmatization is a smarter way of simplifying words while preserving their meaning.

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
punctuations="?:!.,;"
```

```python
sentence_words = nltk.word_tokenize(sentence)
```

```python
sentence_words
```

```python
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)
```

```python
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
